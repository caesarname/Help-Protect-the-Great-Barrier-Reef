```python
def convert(size, box):
    dw = 1./size[0]
    dh = 1./size[1]
    x = (box[0] + box[1])/2.0
    y = (box[2] + box[3])/2.0
    w = box[1] - box[0]
    h = box[3] - box[2]
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return (x,y,w,h)
```

```python
def xyxy2yolo(box, size):
    dw = 1./size[0]
    dh = 1./size[1]
    x = (box[0] + box[1])/2.0
    y = (box[2] + box[3])/2.0
    w = box[1] - box[0]
    h = box[3] - box[2]
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return (x,y,w,h)

def coco2yolo(box,size):
    b = list(map(lambda x: float(x), box))
    x1 = b[0]
    y1 = b[1]
    x2 = x1 + b[2]
    y2 = y1 + b[3]
    return xyxy2yolo([x1,x2,y1,y2],size)


def convert_to_yolo(ann,size = (1280,720)):
  x,y,w,h = ann['x'],ann['y'],ann['width'],ann['height']
  return coco2yolo([x,y,w,h],size)
```

```python
def write_annotation(annotations,path):
  bbox = list(map(convert_to_yolo,annotations))
  with open(path,"w") as f:
    for box in bbox:
      strs = " ".join(["0"] + list(map(str,box)))
      f.write(strs + "\n")
```

```python
# genereate train.txt
from tqdm.auto import tqdm
train_df = df[df['mode'] == "train"]
root = r"/content/train_images"
target= r"/content/dataset/images/train"
target_label = r"/content/dataset/labels/train"
with open("./dataset/train.txt","w") as f:
  for row in tqdm(train_df.values):
    video_id = row[0]
    frame_id = row[2]
    path   = os.path.join(root,f"video_{video_id}",f"{frame_id}.jpg")
    targetimg  = os.path.join(target,f"{frame_id}.jpg")
    shutil.copyfile(path,targetimg)
    path = os.path.join("/content/dataset/images/train",f"{frame_id}.jpg")
    f.write( path + "\n")
    annotations = row[5]
    path = os.path.join(target_label,f"{frame_id}.txt")
    write_annotation(annotations,path)
```

```python
from moviepy.editor import *
import os
import cv2 as cv
from tqdm.auto import tqdm
```

```python
def get_video_clip(root):
  filenames = os.listdir(root)
  filenames = list(filter(lambda x : x.endswith("jpg"),filenames))
  filenames = sorted(filenames,key = lambda x : int(x.split(".")[0]))
  filepaths = list(map(lambda x : os.path.join(root,x),filenames))

  images  = [
      cv.cvtColor(cv.imread(path),cv.COLOR_BGR2RGB)
      for path in tqdm(filepaths)
  ]
  clip = ImageSequenceClip(images,fps = 25).resize(0.5)
  return clip
```
